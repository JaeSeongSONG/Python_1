```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image

img = Image.open('C:\pic\minion.jpg')
img
```

```python
np.array(img)
# 안쪽 리스트의 길이는 행렬의 열의 수 즉, 가로 크기가 되고 바깥쪽 리스트의 길이는 행렬의 행의 수, 즉 세로 크기가 된다
# c = ([[1,2,3], [2,3,4]]) 2 X 3 행렬
# 행의 수 = len(c), 열의 수 = len(c[0])
```

```python
imgArray = np.array(img)
imgArray
result = imgArray*[0,0,1] # RGB 순서
plt.imshow(result)
plt.show()
```

```python
imgArray = np.array(img)
imgArray
result = imgArray*[1,0,0] # RGB 순서
plt.imshow(result)
plt.show()
```

```python
imgArray = np.array(img)
imgArray
result = imgArray*[0,1,0] # RGB 순서
plt.imshow(result)
plt.show()
```
